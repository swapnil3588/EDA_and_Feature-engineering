## EDA and Feature Engineering

```python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python

# importing file

df = pd.read_excel("Data_Train.xlsx")
df.head()
```

```python
# checking data types

df.info()
```

```python

# we will clean data by columns one by one
```

```python
# frist column - Date_of_Journey
```

```python
df["Date"] = df["Date_of_Journey"].str.split("/").str[0].astype(int)
df["Month"] = df["Date_of_Journey"].str.split("/").str[1].astype(int)
df["Year"] = df["Date_of_Journey"].str.split("/").str[2].astype(int)
```

```python
df.info()
```

```python


# now we dont need Date of journey columns ,we can drop that column

df.drop("Date_of_Journey", axis=1, inplace=True)
```

```python

# we will look at our data
df.head(5)
```

```python

# now we need to clean arrival_time and dep_time columns 
# and make new columns for hours and mins and delete original columns
```

```python
# dep_time column

df["Dep_hour"] = df["Dep_Time"].str.split(":").str[0].astype(int)
df["Dep_min"] = df["Dep_Time"].str.split(":").str[1].astype(int)
```

```python

# Arrival_time columns       

df["Arrival_Time"] = df["Arrival_Time"].str.split(" ").str[0]


df["Arrival_hour"] = df["Arrival_Time"].str.split(":").str[0].astype(int)
df["Arrival_min"] = df["Arrival_Time"].str.split(":").str[1].astype(int)
```

```python

# now we will drop both original columns

df.drop(["Arrival_Time","Dep_Time"], axis=1, inplace=True)
```

```python

# now we take look at table

df.head(5)
```

```python

# we will again look at info

df.info()
```

```python

# as wee see above we have cleanded half of our columns
```

```python
# fristly we will drop  Route column , we dont need it
df.drop("Route", axis=1, inplace=True)
```

```python

# now we will drop false value  Duration

df.drop(6474, axis=0, inplace=True)
```

```python
# functions to clean and convert hours to min 

def con_hours(hour):
    h = hour.replace("h","")
    hr = int(h)*60
    return hr

def con_mins(min):
    m = min.replace("m","")
    return int(m)
    
```

```python

# now we will convert hours to min from Duration column

hour = df["Duration"].str.split(" ").str[0].apply(con_hours)
min = df["Duration"].str.split(" ").str[1].fillna("0m").apply(con_mins)


df["Duration"] = hour + min
```

```python
# rename the table
df.rename(columns={"Duration":"Duration_min"},inplace=True)
```

```python

# we will look at table again
df.head()
```

```python
df.info()
```

```python

# we will look at column - Total_stops    (it has 1 null value)
# and convert this column into numberic

df["Total_Stops"].unique()
```

```python
df["Total_Stops"] = df["Total_Stops"].map({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4})
```

```python

# now we will look at 1 null value in Total_stops column
df[df["Total_Stops"].isnull()]
```

```python

# from the data available we can say that this null value can feel with 1
df["Total_Stops"] = df["Total_Stops"].fillna(1)
```

```python

# we will se at info

df.info()
```

```python


# now the remaining four columns have lot of differnt values 
# so we will use labelencoder to convert object values to numberic value

```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df["Airline"] = le.fit_transform(df["Airline"])
df["Destination"] = le.fit_transform(df["Destination"])
df["Source"] = le.fit_transform(df["Source"])
df["Additional_Info"] = le.fit_transform(df["Additional_Info"])
```
